# Python Type Annotations

## Introduction
- Type annotations in Python allow you to specify the expected types of variables, function parameters, return values, and class attributes. 
- They make code more readable, enable early error detection, and enhance tooling support(by understanding the types, IDEs offer intelligent code completion, type inference, and refactoring support.).

## Variable Annotations


```python
x: int = 5
```

This defines that x should be of type int, aiding in readability and allowing tools like Mypy to check for errors

## List Annotations

```python
from typing import List
numbers: List[int] = [1, 2, 3]
```

This specifies that numbers should be a list of integers.

```python
from typing import List

numbers: List[int] = [1.5, 2, 3]
print(numbers)
```
